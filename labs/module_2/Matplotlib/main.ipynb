```python
# ENHANCED VERSION to see more matplotlib features

# Variables
fs = 16 # fontsize
fw = 700 # fontweight
lw = 3 # lineweight
y_rot = 0 # ylabel rotation

font_title = {'family': 'arial',
        'color':  'black',
        'weight': 'normal',
        'size': 24,
        }

font_axis = {'family': 'arial',
        'color':  'black',
        'weight': 'normal',
        'size': 12,
        }

# Create a figure of a fixed size and axes
fig, [ax1, ax2] = plt.subplots(1, 2, figsize = (10,5))
fig.tight_layout(w_pad=5)

# First plot adding a title and x and y labels
ax1.plot(x,y, color='g')
ax1.set_title ('LINEAR', fontdict = font_title)
ax1.set_xlabel('x', fontdict = font_axis)
ax1.set_ylabel('y', fontdict = font_axis)

# Second plot adding a title and x and y labels
ax2.plot(x,z, color = 'b')
ax2.set_title ('QUADRATIC', fontdict = font_title)
ax2.set_xlabel('x', loc = "center", fontdict = font_axis)
ax2.set_ylabel('z', fontdict = font_axis)


# Show the plot
plt.show;
```

#### Plot both y=x^2 and y=exp(x) in the same plot using normal and logarithmic scale.
Hint: Use set_xscale and set_yscale

```python
# Create a figure of a fixed size and axes
fig, [ax1, ax2] = plt.subplots(1, 2, figsize = (8,5))
fig.tight_layout(w_pad=5)


# Normal plot
ax1.plot(x,'r', np.exp(x), 'g')
ax1.set_title("normal scale (y)")

# Logarithmic scale
ax2.plot(x,'r', np.exp(x), 'g')
ax2.set_yscale("log")
ax2.set_title("Log scale (y)")

# Show plot
plt.show();
```

```python
# ALTERNATIVE WITH LEGEND

# Create a figure of a fixed size and axes
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(10,6))

# Normal plot
ax1.plot(x,'r', label='linear')
ax1.plot(np.exp(x), 'g', label='exp')
ax1.set_title("normal")
ax1.legend()

# Logarithmic scale
ax2.plot(x,'r', label = "linear")
ax2.plot(np.exp(x), 'g', label="exp")
ax2.set_yscale("log")
ax2.set_title("logarithmic")
ax2.legend()

# Show plot
plt.show();
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month.
Is Fitbitter more active on weekend or workdays?
All plots must be in the same jupyter notebook cell.

Hints:

- Use Months_encoded and Week_or Weekend columns.
- Use matplolib.pyplot object oriented API.
- Set your size figure to 12,4
- Explore plt.sca
- Explore plt.xticks
- Save your figures

```python
# MEDIAN STEPS BY MONTH_ENCODED
# Find the median steps for each month
mean_steps_month = fitbit['Steps'].groupby(fitbit['Months_encoded']).mean()
median_steps_month = fitbit['Steps'].groupby(fitbit['Months_encoded']).median()

# Create a figure of a fixed size and axes
fig, ax = plt.subplots(1,2, figsize = (10,5))
fig.tight_layout(w_pad=7)

# Set the current axes instance to ax[0] - Now plt methods will affect ax[0]
plt.sca(ax[0])

# # Plot the weekday steps in the current axes
median_steps_month.plot(kind = 'line', color = 'b', alpha = 0.5)

# Add labels, title, etc
plt.style.use('ggplot')
plt.title("Steps per months (median)")
plt.xlabel("")
plt.ylabel("number of steps (median)")
plt.xticks(range(13), ["", "january", "february", "march", "april", "may", "june", "july", "august", 
            "september", "oktober", "november", "december"], rotation=90)


# Save the figure
plt.savefig('figures/median_steps_months.png')

# MEDIAN STEPS BY WORK_OR_WEEKEND
# Set the current axes instance to ax[1] - Now plt methods will affect ax[1]
plt.sca(ax[1])

# Find the median steps for workdays and weekdays
mean_steps_work_weekend = fitbit['Steps'].groupby(fitbit["Work_or_Weekend"]).mean()
median_steps_work_weekend = fitbit['Steps'].groupby(fitbit["Work_or_Weekend"]).median()


# Plot the workday and weekend steps in the current axes
median_steps_work_weekend.plot(kind = "bar", color = 'b', alpha = 0.5)

# Add labels, title, etc
plt.title("Workdays versus weekends")
plt.ylabel("Steps per months")
plt.xlabel("")
plt.xticks(range(2), ["workdays", "weekend"], rotation = 0)

# Save the figure
plt.savefig("figures/work_vs_weekend_steps")


plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

Minutes Lightly Active vs Steps
Minutes Very Active vs Steps
Minutes Sedentary vs Steps

```python
# ALTERNATIVE 1
# Create a df with the columns we are interested in 
df_small = fitbit[["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary", "Steps"]]

# Create a figure of a fixed size and axes
fig, ax = plt.subplots(1,3, figsize =(12,3))
fig.tight_layout(w_pad=7)

# Iterate to draw each scatter plot
for i in range(3):
    plt.sca(ax[i])
    plt.scatter(df_small.iloc[:, i], df_small.iloc[:, 3], color = 'darkgreen', alpha = 0.3)
    plt.xlabel(df_small.columns[i])
    plt.ylabel('number of steps')
    plt.title(f'Steps when: \n {df_small.columns[i]}')
```
