```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms) 

print(x)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

 Your output should look like below:
 
 ![uniform distribution](ud.png)



```python
def func_uniform(bottom, ceiling, count):
    """
    input: float/integer for bottom/ceiling; integer for count
    output: list of random variables
    """
    x = uniform.rvs(size=count)
    a = bottom
    b = ceiling
    randoms  = a + (b-a)*x
    return randoms
```

```python
data_1 = func_uniform(10,15,100)
data_2 = func_uniform(10,60,1000)

fig, axes = plt.subplots(1,2, dpi=150)
plt.tight_layout()

plt.sca(axes[0])
axes[0].set_ylim(bottom=0,top=150)
plt.ylabel("number of observations")
plt.xlabel("random number")
plt.hist(x=data_1)

plt.sca(axes[1])
axes[1].set_ylim(bottom=0,top=150)
plt.ylabel("number of observations")
plt.xlabel("random number")
plt.hist(x=data_2)
plt.show();
```

How are the two distributions different?

```python
# your answer below
"""
first one is range 5, size 100 (so 1:20)
second one is range 50, size 1000 (so also 1:20)

The difference is therefore not in the in the 'fine-ness' of the bars, but in the number of 'samples' per bar: 
you draw way more numbers in the second plot, over a 'wider'range.

"""
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

You can check the expected output

![normal distribution](nd.png)

```python
# your code here

def func_normal(mean, stdev, count):
    """
    input: float/integer for mean, stdev; integer for count
    output: list of random variables
    """
    x = norm.rvs(loc = mean, scale=stdev, size=count)
    return x
```

```python
data_1 = func_normal(10,1,1000)
data_2 = func_normal(10,50,1000)

fig, axes = plt.subplots(1,2, dpi=150)
plt.tight_layout()

plt.sca(axes[0])
axes[0].set_xlim(5,15)
plt.ylabel("number of observations")
plt.xlabel("random number")
plt.hist(data_1, bins = int(np.sqrt(1000)))

plt.sca(axes[1])
axes[1].set_xlim(-200,200)
plt.ylabel("number of observations")
plt.xlabel("random number")
plt.hist(data_2, bins = int(np.sqrt(1000)))

plt.show();
```

How are the two distributions different?

```python
# your answer below
"""
The number of observation is the same, but the range on the x-axis is different (min, max and stdev).
"""
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 1 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)


```python
# your code here
def func_exp(mean, size):
    x = np.random.exponential(mean, size)
    return x
```

```python
data_1 = func_exp(10, 1)
data_2 = func_exp(10, 100)

fig, axes = plt.subplots(1,2, dpi=150)
plt.tight_layout()

plt.sca(axes[0])
plt.ylabel("number of observations")
plt.xlabel("number")
plt.hist(data_1, bins=100)

plt.sca(axes[1])
plt.ylabel("number of observations")
plt.xlabel("number")
plt.hist(data_2, bins=100)
plt.show();
```

How are the two distributions different?

```python
# your answer here
"""
the first one just draws 1 sample from the exponential distribution (hence 1 observation), 
while the other one draws 100 samples from the distribution.
"""
```

## Exponential Distribution in the Bank

Suppose that the amount of time one spends in a bank is exponentially distributed with a mean of 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

my text: 
#### Exponential: mu = stdev;  lambda = rate

CDF: 𝐹(𝐴)=1−𝑒(−𝜆𝑎)

```python
# your answer here
# Hint: This is same as saying P(x<15)

k = 15
mean = 10
lmbd = 1/mean #lambda is rate. eg 1 customer leaves every [time unit]
stdev = mean #exponential function: mean = stdev

#using built in CDF:
y = expon.cdf(x=k, scale=mean) #x, scale (= stdev = mean). DO NOT ENTER LOC!! loc = 0 in expon.cdf


#using formula:
F_K = 1-math.exp(-(lmbd)*(k))

print("The chance that someones leaves the bank within 15 minutes is:", round((y*100),2), "%",
      "\nor, according to the formula:", round((F_K*100),2), "%")
```

What is the probability that the customer will spend more than 15 minutes

```python
# your answer here
print("The chance that someones leaves the bank within 15 minutes is:", round(((1-y)*100),1), "%",
    "\nor, according to the formula:", round(((1-F_K)*100),2), "%")
```

A delivery company needs 35 minutes to deliver a package, with a standard deviation of 8 minutes. Suppose that in one day, they deliver 200 packages.

**Hint**: `stats.norm.cdf` can help you find the answers.

#### Step 1: What is the probability that the mean delivery time today is between 30 and 35 minutes?

```python
# your code here

stdev = 8 #scale
mean = 35 #loc
n = 200 #size

margin_of_error = stdev / np.sqrt(n) #corrects for sample size

#what is the chance that we sample from the true_dist and find a mean between 30 or 35?
obs_mean_1 = 30
obs_mean_2 = 35

print(round((norm.cdf(x= obs_mean_2, loc=mean, scale=margin_of_error) - norm.cdf(x= obs_mean_1, loc=mean, scale=margin_of_error))*100, 1), "%")

#what happens if you use t instead of norm:
SEM = stdev / np.sqrt(n)
t_try = (t.cdf(x= obs_mean_2, loc=mean, scale=SEM, df=n-1)- t.cdf(x= obs_mean_1, loc=mean, scale=SEM, df=n-1))*100

print(t_try)
```

```python
#see in more detail how the upper/lower values of norm and t relate
upper_norm = norm.cdf(x= obs_mean_2, loc=mean, scale=SEM)
lower_norm = norm.cdf(x= obs_mean_1, loc=mean, scale=SEM)
upper_t = t.cdf(x= obs_mean_2, loc=mean, scale=SEM, df=n-1)
lower_t = t.cdf(x= obs_mean_1, loc=mean, scale=SEM, df=n-1)
upper_t2 = t.cdf(x= obs_mean_2, loc=mean, df=n-1)
lower_t2 = t.cdf(x= obs_mean_1, loc=mean, df=n-1)


print(upper_norm, '\n',upper_t, '\n',upper_t2, '\n\n',lower_norm, '\n',lower_t,'\n',lower_t2)
```
